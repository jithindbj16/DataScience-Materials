```python
def jumpingOnClouds(c):
    count = 0
    i = 0

    while (True):
            if ((i+2) < len(c) and c[i + 2] == 0):
                 i=i+2
            elif ((i+1) < len(c)):
                i=i+1
            else:
                break
            count=count+1
    return count
    

n = int(input())
c = list(map(int, input().rstrip().split()))
result = jumpingOnClouds(c)
result
```
