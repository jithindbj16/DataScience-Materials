# 
Hypothesis:

    H0: Mean of Mumbai = Mean of Bangalore
    HA: Mean of Mumbai !=  Mean of Bangalore

```python
import pandas            as pd
import numpy             as np
import matplotlib.pyplot as plt
import seaborn           as sns
```

```python
df=pd.read_csv("Case+Study+Hotel+Tariff+Comparison (1).csv")
df.head()
```

```python
l1=list(df.iloc[:,1])
l2=list(df.iloc[:,2])
```

```python
print('Count, Mean and standard deviation of Mumbai: %3d, %3.2f and %3.2f' % (len(l1), np.mean(pd.DataFrame(l1).dropna()),np.std(pd.DataFrame(l1),ddof =1)))
print('Count, Mean and standard deviation of Bangalore: %3d, %3.2f and %3.2f' % (pd.DataFrame(l2).count(), np.mean(pd.DataFrame(l2).dropna()),np.std(pd.DataFrame(l2),ddof =1)))
```

```python
dfn = pd.DataFrame()

df1 = pd.DataFrame({'City': 'Bangalore', 'Price':l2})
df2 = pd.DataFrame({'City': 'Mumbai', 'Price':l1})         

dfn = dfn.append(df2) 
dfn = dfn.append(df1)
dfn
```

```python
import scipy.stats             as     stats
```

```python
NofSample=2
dfd=(len(df1)+len(df2))- NofSample      # df Denominator
dfn= NofSample - 1                                     # df Numerator
crit = stats.f.ppf(q = 1-0.05,dfn=dfn, dfd= dfd)
print('F critical value with .95 confidence %3.2f' %crit)
```

```python
p_value = 1 -stats.f.cdf(0.497075, dfn = dfn, dfd = dfd)
print('P value  for the calculated F value %3.5f' % p_value)
```

```python
import statsmodels.api         as     sm
from   statsmodels.formula.api import ols
 
mod = ols('Price ~ City', data = dfn).fit()
aov_table = sm.stats.anova_lm(mod, typ=2)
print(aov_table)
```

# As our F-Statistic value is much higher than our F-Critical value, we reject the null hypothesis. Also, the p-value is much lesser than 0.05.

Hence, the means of two samples are not equal.

Bangalore is more Expensive than Mumbai
