```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
x=data[['lstat']]
y=data[['medv']]
```

```python
pl=PolynomialFeatures(degree=6)               # quadratic regresssion model
```

```python
qr=pl.fit_transform(x)
```

```python
xqr=pd.DataFrame(qr)
xqr.head()
```

```python
xqr=xqr.drop(columns=0)
```

```python
xqr.tail()
```

```python
xqr.info()
```

```python
idx=np.arange(data.shape[0])
```

```python
y.index=idx
```

```python
xqr=pd.concat([xqr,y],axis=1)
```

```python
xqr.shape
```

```python
xqr.columns=['f1','f2','f3','f4','f5','f6','Medv']
```

```python
m2=sm.ols('Medv~f1+f2+f3+f4+f5+f6',xqr).fit()
```
